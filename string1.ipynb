```python
str="Today is 19-Jan-2026 MONDAY"
B=V=C=D=O=0
for i in str:
  if i ==" ":
    B=B+1
  elif i in "aeiou":
    V=V+1
  elif i.isalpha():
    C=C+1
  elif i.isdigit():
    D=D+1
  else:O=O+1
print("vowels",V)
print("blanks",B)
print("consonant",C)
print("digits",D)
print("others",O)
```

```python
#extracting all string with prefix
codes=["CRP_RICE_2026",
       "CRP_WHEAT_2025",
       "CRP_MANGO_2026",
       "CRP_MAIZE_2024"]
filtered=[]
for crop_code in codes:
  if crop_code.startswith("CRP_"):
    filtered.append(crop_code)
print(filtered)
```

```python
#extract and clean in one go
codes=["CRP_RICE_2026",
       "CRP_WHEAT_2025",
       "CRP_MANGO_2026",
       "CRP_MAIZE_2024"
      ]
clean_codes=[]
for code in codes:
  if code.startswith("CRP_"):
    clean_codes.append(code[4:])
print(clean_codes)
```
