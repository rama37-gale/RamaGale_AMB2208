```python
import pandas as pd
data={
    "farm":["farm a","farm b","farm c","farm d"],
    "Crop":["wheat","rice","wheat","maizze"],
    "areahect":[10,15,8,12],
    "prodtons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
#farms with yeld geater than 3
high_yield=df[df["yield_tos_hec"]>3]
print(high_yield)
```

```python
#sort farm by highest production
sorted_df=df.sort_values(by="prodtons",ascending=False)
print(sorted_df)
```

```python
import pandas as pd
import numpy as np

#create data with missing values
data=pd.DataFrame({
    "plot":["A","B","C","D","E"],
    "yields_kg": [2500,np.nan,2800,np.nan,2600],
    "rainfall": [150,200,np.nan,180,170],
    "tempaerature":[28,30,29,np.nan,27]
})
print()
print("===data with missing values===")
print(data)
#detecting missing data
print()
print("===detecting missing data===")
print("is null?")
print(data.isnull())
print()
print(f"total missing per coloumn:")
print(data.isnull().sum())
```

```python
# filling missing values
print("filling strategies")
print()
# fill with mean
data_mean=data.copy()
print(data_mean["yields_kg"])
data_mean["yields_kg"]=data_mean["yields_kg"].fillna(data_mean["yields_kg"].mean())
print("fill with mean")
print(data_mean["yields_kg"])

```

```python
#fill with median
data_median=data.copy()
data_median["rainfall"]=data_median["rainfall"].fillna(data_median["rainfall"].median())
print("fill with median")
print(data_median["rainfall"])
#
```

```python
#forward fill (use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method="ffill")
print("forward fill")
print(data_ffill)
#ffill-it replaces missing values(nan)
# with the last valid (non-missing) value above it in the coloumn.
```

```python
# dropping missing data
print("droping rows")
data_dropped=data.dropna()
print(data_dropped)
print(f"original rows:{len(data)},After drop:{len(data_dropped)}")
```
