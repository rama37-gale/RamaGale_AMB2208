```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv('Salary_Data.csv')
dataset.head()
```

```python
#slicing the data set
x=dataset.iloc[:,0:1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

```python
#splitting the data setinto training set testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
print(dataset.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#calculate the mean values of input and output
import numpy as np
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean,y_mean)
```

```python
#calculate intercept and coefficient
n=len(x_train)
num,den=0,0
for i in range(n):
   num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
   den+=(x_train[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("b0:",b0)
```

```python
#predicting result on test dataset
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
#mean square error
n1=len(y_test)
mse=0
for i in range(n1):
  mse+=(y_test[i]-y_pred_code[i])**2
mse=mse/n1
print(mse)
```

```python
#root mean squar error
rmse=np.sqrt(mse)
print(rmse)
```

```python
#create an object of the algorithm
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
# trainthe model
reg.fit(x_train,y_train)
```

```python
#predict model on testing dataset
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
#calculate intercept and coefficient
print(reg.intercept_)
print(reg.coef_)
```

```python
#evaluation metrc
from sklearn import metrics
print('mean absolute error:',metrics.mean_absolute_error(y_test,y_pred))
print('mean squared error:',metrics.mean_squared_error(y_test,y_pred))
print('root mean squared error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
