```python
from typing import Counter
import random
inventary={}
total_money=0
#fish and prices
fish_prices={
    "rohu":23,
    "catla":18,
    "sinchu":25
}
def showmenu():
    print("\n-------fishing game menu-----")
    print("1 catch fish")
    print("2 view inventary")
    print("3 sell fish")
    print("4 exit")
def catchfish():
    fish=random.choice(list(fish_prices.keys()))
    inventary[fish]=inventary.get(fish,0)+1
    print(f"you caughht a {fish}")
def viewfish():
   if inventary=={}:
    print("no fish")
   else:
    print("your inventary.")
    for fish, count in inventary.items():
      print(f"{fish}:{count}")
def sellfish():
  global total_money
  if inventary=={}:
    print("no fish to sell")
  else:
    earning=0
    for fish,count in inventary.items():
      earning+=fish_prices[fish]* count

    total_money+= earning
    inventary.clear()
    print(f"you sold all fish for ${earning}!")
def exit():
  print(f"final money earned:${total_money}")
  print("thanks for playing")










showmenu()
#game loop
while True:
  ch=int (input("choose an option."))
 #catch fish
  if ch==1:
     catchfish()

  #view inventery
  elif ch==2:
    viewfish()
      #sell fish
  elif ch==3:
    sellfish()

  #exit
  elif ch==4:
     exit()
     break
else:
  print("invalid choice ,try again")













```
