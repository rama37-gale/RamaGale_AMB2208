```python
#8/2/2026
import numpy as np
crop_yi=np.array([[12.5,13.2,14.1,13.8,15.2],
    [10.3,11.5,10.8,12.1,11.9],
   [15.1,14.3,12.4,10.4,11.5],
   [11.2,10.9,12.1,11.7,12.8]])
print("crop yield :")
print(crop_yi)
print(f'\nshape: {crop_yi.shape}')
print(f'\ndata type{crop_yi.dtype}')
print(f'totsl element:{crop_yi.size}')
print(f'number of dimension:{crop_yi.ndim}')
```

```python
total=np.sum(crop_yi)
ave_farm=np.mean(crop_yi,axis=1)
ave_sea=np.mean(crop_yi,axis=0)
max=np.max(crop_yi)
min=np.min(crop_yi)
print(f'total tield{total}')
print(f'{ave_farm}')
print(f"{ave_sea}")
print(max)
print(min)
```

```python
#finding best performance
best_farm=np.argmax(ave_farm)
print("\nbest performing farm:farm",best_farm+1)
print("there average yield",round(ave_farm[best_farm],2),"tons")

```
